# Gold Price Analysis (2014–2026)
## Exploratory Data Analysis (EDA)

This notebook performs exploratory data analysis on historical gold price data including Open, High, Low, Close (Price), Volume, and percentage change.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

## Load Dataset

```python
df = pd.read_csv('Gold Price.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
df.head()
```

# 3.1 Exploratory Data Analysis

## 3.1.1 Date, Open & Price Analysis

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date', y='Price', color='blue')
plt.title('Historical Price Trends of Gold (2014–2026)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date', y='Open', color='orange')
plt.title('Daily Opening Price Fluctuations (2014–2026)')
plt.xlabel('Date')
plt.ylabel('Open')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Open'], label='Open', alpha=0.7)
plt.plot(df['Date'], df['Price'], label='Price', alpha=0.5)
plt.title('Comparison of Open and Price Trajectories (2014–2026)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sns.regplot(data=df, x='Open', y='Price',
            scatter_kws={'alpha':0.3, 's':10},
            line_kws={'color':'red'})
plt.title('Correlation between Open and Price')
plt.xlabel('Open')
plt.ylabel('Price')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

## 3.1.2 High & Low Price Analysis

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date', y='High', label='High')
plt.title('Daily Highest Price of Gold (2014–2026)')
plt.xlabel('Date')
plt.ylabel('High Price')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date', y='Low', color='red', label='Low')
plt.title('Daily Lowest Price of Gold (2014–2026)')
plt.xlabel('Date')
plt.ylabel('Low Price')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

## 3.1.3 Volume & Percentage Change Analysis

```python
plt.figure(figsize=(12, 5))
plt.plot(df['Date'], df['Volume'])
plt.title('Gold Trading Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
sns.histplot(df['Chg%'], bins=30)
plt.title('Distribution of Gold Price Percentage Change')
plt.xlabel('Percentage Change')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.scatter(df['Volume'], df['Price'])
plt.title('Relationship Between Volume and Gold Price')
plt.xlabel('Volume')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

# 3.2 Price Analysis

```python
print('Summary Statistics of Gold Price (2014–2026)')
summary = df['Price'].describe()
summary
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df['Date'], df['Price'], color='gold')
plt.title('Daily Gold Price Trend (2014–2026)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Price'], bins=30, color='orange', edgecolor='black')
plt.title('Distribution of Daily Gold Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
min_price = df['Price'].min()
max_price = df['Price'].max()

plt.figure(figsize=(6, 4))
plt.bar(['Minimum Price', 'Maximum Price'], [min_price, max_price], color=['green', 'red'])
plt.title('Minimum and Maximum Daily Gold Prices')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```
